# Image blur and smoothing techniques

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import cv2
```

```python
img=cv2.imread("dog.png")
```

```python
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
```

```python
plt.imshow(img)
```

# Averaging Blur

```python
img1=cv2.blur(img,(3,3))
plt.imshow(img1)
```

```python
img1=cv2.blur(img,(7,7))
plt.imshow(img1)
```

# Gaussian Blur

```python
img2=cv2.GaussianBlur(img,(5,5),0)
plt.imshow(img2)
```

```python
img2=cv2.GaussianBlur(img,(7,7),0)
plt.imshow(img2)
```

# Median Blur

```python
img3=cv2.medianBlur(img,5)
plt.imshow(img3)
```

# Bilateral Filter

```python
img3=cv2.bilateralFilter(img,9,75,75)
plt.imshow(img3)
```

```python
img3=cv2.bilateralFilter(img,11,175,175)
plt.imshow(img3)
```
