# Blending and Pasting

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import cv2
```

```python
img1=cv2.imread("lion.png")
img2=cv2.imread("cat.png")
```

```python
img1.shape
```

```python
img2.shape
```

```python
img1 = cv2.cvtColor(img1, cv2.COLOR_BGR2RGB)
img2 = cv2.cvtColor(img2, cv2.COLOR_BGR2RGB)
```

```python
img1 =cv2.resize(img1,(2000,1300))
img2 =cv2.resize(img2,(2000,1300))
```

```python
plt.imshow(img1)
```

```python
img2.shape
```

```python
img1.shape
```

# Blending the Images

```python
blending=cv2.addWeighted(src1=img1,alpha=0.6,src2=img2,beta=0.5,gamma=0.6)
```

```python
plt.imshow(blending)
```

# Pasting the Images

```python
img1=cv2.imread("lion.png")
img2=cv2.imread("cat.png")
img1 =cv2.resize(img1,(1000,600))
img2=cv2.resize(img2,(450,300))
```

```python
img1 = cv2.cvtColor(img1, cv2.COLOR_BGR2RGB)
img2 = cv2.cvtColor(img2, cv2.COLOR_BGR2RGB)
```

```python
lion_img = img1
cat_img=img2
```

```python
x_offset = 0
y_offset = 0
```

```python
lion_img[y_offset:y_offset+cat_img.shape[0], x_offset:x_offset+cat_img.shape[1]] = cat_img
```

```python
plt.imshow(lion_img)
```
